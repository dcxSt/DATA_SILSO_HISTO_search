```python
# database connector

def database_connector(the_database="DATA_SILSO_HISTO"):
    try:
        mydb = mysql.connector.connect(
            host="localhost",
            user="root",
            database=the_database
        )

        cursor = mydb.cursor()
        cursor.prepared=True # this is useless?
        
        print("\nmysql connection sucessfully established...")
        
        return cursor,mydb

    except Exception as e:
        print("\nAn error has occured connecting to mysql database")
        raise Exception(e)
        
def get_cursor(cursor=None):
    if not cursor:
        cursor,mydb=database_connector()
        return cursor
    else:
        return cursor

# define a cursor object
cursor,mydb = database_connector()
print(cursor)
print(mydb)

# haven't yet tested this to see if it works
def close_database_connection(mydb):
    if mydb.is_connected():
        mydb.close()
        print("closing MySQL connection...")
        return mydb
    else:
        print("\nerror")
        


#close_database_connection(mydb)
```

```python
# select DATA from RUBRIC NUMBER
display=False
def select_data_from_rubric(cursor=None,rubricNumber=710):
    cursor=get_cursor(cursor)
    query = "SELECT * FROM DATA d WHERE d.FK_RUBRICS=(SELECT r.RUBRICS_ID FROM RUBRICS r WHERE r.RUBRICS_NUMBER = "+str(rubricNumber)+")"
    
    cursor.execute(query,params=())
    data = cursor.fetchall()
    return data

def display_select_data_from_rubric(cursor,rubricNumber=710):
    data=select_data_from_rubric(cursor,rubricNumber)
    print("Rubric Number = "+str(rubricNumber))
    print("Size of data : len(data) = ",len(data))
    print("note, FK_RUBRICS corresponds to a different rubrics\
    number than in mitteleugen journals\n")
    print("ID ; DATE ; FK_RUBRICS ; FK_OBSERVERS ; GROUPS ; SUNSPOTS ;\
    WOLF ; QUALITY ; DATE_INSERT ; COMMENT ")
    
    for i in data:
        if i[8]=="":
            print ((i[0]),(i[1]),"  ",(i[2]),"\t",(i[3]),"\t",
                (i[4]),"\t",(i[5]),"\t    ",(i[6]),(i[7]),(i[9]),"\tNa")
        else:
            print ((i[0]),(i[1]),"  ",(i[2]),"\t",(i[3]),"\t",
                (i[4]),"\t",(i[5]),"\t    ",(i[6]),(i[7]),(i[9]),(i[8]))
    
    
if display:
    display_select_data_from_rubric(cursor,rubricNumber=710)
```

```python
# select OBSERVER from OBSERVER ALIAS
display=False
def get_observer_list(cursor):
    # returns a list of the observers (aliases and name etc.)
    query = "SELECT * FROM OBSERVERS"
    cursor.execute(query,params=())
    observers_list=cursor.fetchall()
    return observers_list

def get_observer_aliases(cursor):
    # returns a list of observer aliases
    aliases=[]
    for observer in get_observer_list(cursor):
        aliases.append(observer[1])
    return aliases

def get_observer_first_names(cursor):
    # returns a list of observer first names
    first_names=[]
    for observer in get_observer_list(cursor):
        first_names.append(observer[2])
    return first_names
        
def get_observer_last_names(cursor):
    # returns a list of observer last names
    last_names=[]
    for observer in get_observer_list(cursor):
        last_names.append(observer[3])
    return last_names
    
    
    
def select_data_from_observer(cursor=None,alias="%franz%"):
    cursor = get_cursor(cursor)
    query = "SELECT * FROM OBSERVERS o WHERE o.ALIAS like"

```
