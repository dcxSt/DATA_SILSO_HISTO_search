```python
# unpickle kew_penumbra as kew_data
f=open("pickles/kew_penumbra.pickle","rb")
kew_data = pickle.load(f)
f.close()

# change kew's data according to formula derived in rub 306
a = 0.763
b = 0.042
kew_derived=[]
for i in kew_data:
    j = [i[0],i[1],1011,i[3]]
    groups = i[4]
    penumbra = i[5]
    # r = a * (10 g + b * f)
    wolf = int(a * (10*groups + b*penumbra)+0.5)# round to nearest integer
    sunspots = wolf - 10*groups
    if sunspots <0:
        if groups>0:
            sunspots=1
    if groups==0:
        sunspots=0
    j.append(groups)
    j.append(sunspots)
    j.append(wolf)
    j.append(i[7])
    j.append('derived')
    j.append(i[9])
    j.append(7)# set the flag to 7
    kew_derived.append(tuple(j))
```
