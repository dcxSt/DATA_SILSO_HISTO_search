```python
# plotting sunspots for 'Kew' and 'Carrington'
different_data=[]
for alias in 'Carrington','Kew','Jenzer','Leppig':
    date=[]
    sunspots_by_date={}
    plotting_dictionary={}
    for i in obs_alias_dictionary[alias]:
        date.append(i[1])
        sunspots_by_date[i[1]]=i[3]
    date.sort()
    sunspots=[]
    for i in date:
        if sunspots_by_date[i]<6000:
            sunspots.append(sunspots_by_date[i])
        else:
            sunspots.append(0)
            for j in date:
                if j==i:
                    print(j)
    
    different_data.append([date,sunspots,alias])
    
plt.figure(figsize=(15,7))
plt.title("Carrington, Kew, Jenzer and Leppig")
plt.ylabel("Sunspots seen (not the wolf number)")
plt.xlabel("Time: in years AD")
plt.plot(different_data[0][0],different_data[0][1],"rx",label='Carrington - superhuman eyesight')
plt.plot(different_data[1][0],different_data[1][1],"ro",label="Kew - superhuman eyesight")
plt.plot(different_data[2][0],different_data[2][1],"bx",label='Jenzer - regular eyesight')
plt.plot(different_data[3][0],different_data[3][1],"bo",label='Leppig - regular eyesight')
plt.legend()
plt.show()
    
```

```python
# plotting sunspots for others
date=[]
sunspots_by_date={}
plotting_dictionary={}
for alias in 'Kew','Carrington':
    for i in obs_alias_dictionary[alias]:
        date.append(i[1])
        sunspots_by_date[i[1]]=i[3]
    date.sort()
    sunspots=[]
    for i in date:
        if sunspots_by_date[i]<40000:
            sunspots.append(sunspots_by_date[i])
        else:
            sunspots.append(0)
            for j in date:
                if j==i:
                    print(j)
        
    plt.figure(figsize=(20,10))
    plt.title(alias)
    plt.plot(date,sunspots,"rx")
    plt.show()
```

```python
# plotting wolf
for alias in obs_alias_dictionary:
    #we make a dictionary key=date --> value=wolf, so that we can then sort it
    date=[]
    wolf_by_date={}
    for i in obs_alias_dictionary[alias]:
        date.append(i[1])
        wolf_by_date[i[1]]=i[4]
    date.sort()
    wolf=[]
    for i in date:
        wolf.append(wolf_by_date[i])
    if alias=="Konkoly" or alias=="Koukoly":
        plt.figure(figsize=(15,7))
        plt.title(alias)
        plt.plot(date,wolf,"x")
        plt.show()
        print()
        




```
