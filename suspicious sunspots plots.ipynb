```python
import numpy as np
import matplotlib.pyplot as plt
import graphs_helper
import time
```

```python
# define a list of suspicious observers with suspicious sunpots
suspicious_observers=["Carrington","Key","Kew","Brunner","Buser",
                     "Sykora","Broger","Guillaume","Heimen",
                     "Quimby","Wolfer","Hrase","Tacchini",
                     "Collins- Haverford","Schwab","Lewitzky",
                     "Weber","Meyer et Billwiller","Schmitd"]
manualy_cleared_observers=["Weber"]
```

```python
# get dictionary from the python file and refine it into 4 
# sections, for non-flagged, flag1 and flag4 and flag5 
data_by_obs = graphs_helper.data_by_obs_alias_good()
data_by_obs_seperate_flags =  {}
for observer in data_by_obs:
    data_by_obs_seperate_flags[observer]=[[],[],[],[]]
    for i in data_by_obs[observer]:
        flag = i[15]
        
        if flag==1:
            data_by_obs_seperate_flags[observer][1].append(i)
        elif flag==4:
            data_by_obs_seperate_flags[observer][2].append(i)
        elif flag==5:
            data_by_obs_seperate_flags[observer][3].append(i)
        else:
            data_by_obs_seperate_flags[observer][0].append(i)
```

```python
# plot sunspots for all those who are suspicious
def display_seperate_flags(observer,interval=None):
    if interval:
        low=time.strftime(interval[0])
        high=time.strftime(interval[1])
    
    plt.figure(figsize=(15,8))
    plt.title(observer)
    plt.xlabel("Date")
    plt.ylabel("Sunspots number")
    
    flag0 = data_by_obs_seperate_flags[observer][0]
    flag1 = data_by_obs_seperate_flags[observer][1]
    flag4 = data_by_obs_seperate_flags[observer][2]
    flag5 = data_by_obs_seperate_flags[observer][3]

    x0=[]
    y0=[]
    
    for i in flag0:
        date=i[1]
        sunspots=i[3]
        if interval:
            date=time.strftime(str(i[1]))
            if date <= high and date >= low:
                x0.append(i[1]) # python spazes out if i use the python date syntax
                y0.append(sunspots)
        else:
            x0.append(i[1])
            y0.append(sunspots)

    x1=[]
    y1=[]
    for i in flag1:
        date=i[1]
        sunspots=i[3]
        if interval:
            date=time.strftime(str(i[1]))
            if date <= high and date>= low:
                x1.append(i[1])
                y1.append(sunspots)
        else:
            x1.append(i[1])
            y1.append(sunspots)
        

    x4=[]
    y4=[]
    for i in flag4:
        date=i[1]
        sunspots=i[3]
        if interval:
            date=time.strftime(str(i[1]))
            if date<=high and date>=low:
                x4.append(i[1])
                y4.append(sunspots)     
        else:
            x4.append(i[1])
            y4.append(sunspots)
        
        

    x5=[]
    y5=[]
    for i in flag5:
        date=i[1]
        sunspots=i[3]
        if interval:
            date=time.strftime(str(i[1]))
            if date<=high and date>=low:
                x5.append(i[1])
                y5.append(sunspots)
        else:
            x5.append(i[1])
            y5.append(sunspots)
            
    plt.plot(x0,y0,"x",label=observer+" flag=0")
    plt.plot(x1,y1,"x",label=observer+" flag=1")
    plt.plot(x4,y4,"x",label=observer+" flag=4")
    plt.plot(x5,y5,"x",label=observer+" flag=5")
    #plt.grid(b=True,color="black",which="major")
    plt.grid()
    if interval:
        plt.xticks(np.linspace(start=low,stop=high,10))
    plt.legend()
    plt.show()

"""
for observer in data_by_obs_seperate_flags:
    if observer in suspicious_observers and observer not in manualy_cleared_observers:
        # plot the data colour coded
        display_seperate_flags(observer)
        """
        
```
